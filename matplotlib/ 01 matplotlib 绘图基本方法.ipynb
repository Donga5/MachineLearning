# matplotlib的基本方法

| 方法名                           | 说明                           |

| title()                          | 设置图表的名称                 |
| xlabel()                         | 设置x轴名称                    |
| ylabel()                         | 设置y轴名称                    |
| xticks(ticks,label,rotation)         | 设置x轴的刻度,rotation旋转角度 |

```python
from matplotlib import pyplot as plt
import numpy as np
x=np.arange(-50,50)
y=x**2
```

### 4. 设置x轴和y轴的刻度

`matplotlib.pyplot.xticks(ticks=None, labels=None, **kwargs)`

`matplotlib.pyplot.yticks(ticks=None, labels=None, **kwargs)`

- ticks: 此参数是xtick位置的列表。和一个可选参数。如果将一个空列表作为参数传递，则它将删除所有xticks
- labels: 此参数包含放置在给定刻度线位置的标签。它是一个可选参数。，
- **kwargs:此参数是文本属性，用于控制标签的外观
 - rotation:旋转角度 如:rotation=45
 - color:颜色  如:color="red"

> <font color="red">xticks到底有什么用，其实就是想把坐标轴变成自己想要的样子</font>

```python
# 日期
times = np.arange(1990,2020).astype(np.str_)
# 销量
sales = np.random.randint(500,2000,size=len(times))
# 绘图,字符型的数据会range()控制times 按照索引哪个进行显示.   labels可以控制具体显示的内容
plt.xticks(range(0,len(times),2),['%s年'%i for i in times],rotation=45)
plt.plot(times,sales)
```

```python
# 每个时间点的销量绘图,如果x轴的数据是字符串,会将数据全部展示
times = ['2015/6/26', '2015/8/1', '2015/9/6', '2015/10/12', '2015/11/17','2015/12/23','2016/1/28','2016/3/4','2016/4/9',
'2016/5/15','2016/6/20','2016/7/26','2016/8/31','2016/10/6','2016/11/11','2016/12/17']

# 随机出销量
sales =np.random.randint(500,2000,size=len(times))

# 绘制图形
plt.plot(times,sales)
```

```python
# 每个时间点的销量绘图,如果x轴的数据是字符串,会将数据全部展示
times = ['2015/6/26', '2015/8/1', '2015/9/6', '2015/10/12', '2015/11/17','2015/12/23','2016/1/28','2016/3/4','2016/4/9',
'2016/5/15','2016/6/20','2016/7/26','2016/8/31','2016/10/6','2016/11/11','2016/12/17']

# 随机出销量
sales =np.random.randint(500,2000,size=len(times))

'''
 默认的刻度是全部都进行显示，xticks可以设置根据原始的刻度如何进行显示
  1.第一个参数控制原有轴上需要显示哪些数据
  2.第二个参数控制在具体的参数上显示的内容
  3.rotation 设置标签的旋转角度，
'''
#plt.xticks(ticks=range(0,len(times),2),rotation=45)
plt.xticks(ticks=range(0,len(times)-1,3),rotation=-30)
print(len(times))

# plt.xticks(ticks=range(0,len(times),2),labels=[i for i in range(16)])
# plt.xticks(ticks=range(0,len(times),2),labels=[i for i in range(4)]) # 当标签信息不足的时候，其余的显示为空

plt.plot(times,sales)
```

#### 举例

```python

x=np.arange(-10,10)
y=x**2

# 有中文的时候先设置参数
plt.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.unicode_minus']=False # 使得负号正常显示
# 设置标题
plt.title('y=x^2 取值范围为-10~10',fontsize=20)

plt.xlabel('x',fontsize=24)
plt.ylabel('y')


# 设置x的刻度
x_ticks=[-10,-5,0,1,2,3]
# 使用列表推导式来自定义标签
x_labels=[f'{x_ticks[i]}cm' for i in range(len(x_ticks))]

plt.xticks(ticks=x_ticks,labels=x_labels)

# linewidth 设置线条的粗细
plt.plot(x,y,linewidth=2)

y1=x
plt.plot(x,y1,linewidth=2)
```

### 6.图例

```python
# 每个时间点的销量绘图
times = ['2015/6/26', '2015/8/1', '2015/9/6', '2015/10/12', '2015/11/17','2015/12/23','2016/1/28','2016/3/4','2016/4/9',
'2016/5/15','2016/6/20','2016/7/26','2016/8/31','2016/10/6','2016/11/11','2016/12/17']

# 随机出收入
income =np.random.randint(500,2000,size=len(times))
# 支出
expenses =np.random.randint(300,1500,size=len(times))

# 绘制图形
plt.xticks(range(1,len(times),2),rotation=45)

# 调用图例时候给线设置标签.label需要使用关键字传参
plt.plot(times,income,label='收入')
plt.plot(times,expenses,label='支出')

# 使用图例,会自动寻找位置
plt.legend()
plt.legend(loc='center left')
```

####  图例的图例位置设置

- loc代表了图例在整个坐标轴平面中的位置（一般选取'best'这个参数值）
 - 第一种:默认是"best",图例自动‘安家’在一个坐标面内的数据图表最少的位置
 - 第二种: loc = 'XXX' 分别有0: ‘best’ （自动寻找最好的位置）
 
| 位置字符串     | 位置值 | 备注               |
| :-------------- | :------ | :------------------ |
| "best"         | 0      | 自动寻找最好的位置 |
| "upper right"  | 1      | 右上角             |
| "upper left"   | 2      | 左上角             |
| "lower left"   | 3      | 左下角             |
| "lower right"  | 4      | 右下角             |
| "right"        | 5      | 右边中间           |
| "center left"  | 6      | 左边中间           |
| "center right" | 7      | 右边中间           |
| "lower center" | 8      | 中间最下面         |
| "upper center" | 9      | 中间最上面         |
| "center"       | 10     | 正中心             |


    


### 7.显示每条数据的值 x,y值的位置

`plt.text(x,y,  string,  fontsize=15,  verticalalignment="top",  horizontalalignment="right") `


- x,y:表示坐标值上的值
- string:表示说明文字
- fontsize:表示字体大小
- verticalalignment：(va)垂直对齐方式 ，参数：[ ‘center’ | ‘top’ | ‘bottom’ | ‘baseline’ ]
- horizontalalignment：(ha)水平对齐方式 ，参数：[ ‘center’ | ‘right’ | ‘left’ ]

```python
# 每个时间点的销量绘图
times = ['2015/6/26', '2015/8/1', '2015/9/6', '2015/10/12', '2015/11/17','2015/12/23','2016/1/28','2016/3/4','2016/4/9',
'2016/5/15','2016/6/20','2016/7/26','2016/8/31','2016/10/6','2016/11/11','2016/12/17']

# 随机出收入
income =np.random.randint(500,2000,size=len(times))

# 绘制图形
plt.xticks(range(1,len(times),2),rotation=45)

# 调用图例时候给线设置标签.label需要使用关键字传参
plt.plot(times,income,label='收入')

# 用来显示某个点的具体信息
# plt.text('2015/10/12',income[3],'最小值点',fontsize=20)

# 利用循环可以显示其他信息, 利用zip进行压缩
for x,y in zip(times,income):
    plt.text(x,y,f'({x},{y})')
```
