```python
movie = ['', '', '']


# ,np.array
real_day1 = np.array([4053, 2548, 1543])

real_day2 = np.array([7840, 4013, 2421])

real_day3 = np.array([8080, 3673, 1342])

y=np.arange(3)


height=0.2
#  
plt.barh(y,real_day1,facecolor='blue',height=height)
plt.barh(y+height,real_day2,height=height,facecolor='yellow')
plt.barh(y+2*height,real_day3,height=height,facecolor='red')

# y
plt.yticks(y+height,movie)

# 
for i in range(3):
    plt.text(real_day1[i],y[i],s=real_day1[i],va='center',ha='left')
    plt.text(real_day2[i],y[i]+height,s=real_day1[i],va='center',ha='left')
    plt.text(real_day3[i],y[i]+2*height,s=real_day1[i],va='center',ha='left')

```
