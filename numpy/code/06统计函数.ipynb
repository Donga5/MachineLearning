```python
import numpy as np 
```

### 平均数

```python
a=np.arange(20).reshape((4,5)) 
print(a)
print(a.mean())
print(a.mean(axis=0))# 计算各列的平均值，得到的是一维数组,axis表示轴
print(a.mean(axis=1))# 计算各行的平均值
```

### 中位数

```python
ar1=np.array([1,2,3,4,5])
ar2=np.array([1,2,3,4,5,6])
print(np.median(ar1))
print(np.median(ar2))
```

### 标准差和方差

```python
ar1=np.array([95,85,75,65,55,45])
ar2=np.array([73,73,71,69,68,67])
print(np.std(ar1))# 标准差
print(np.std(ar2))
print(np.var(ar1))# 方差
print(np.var(ar2))
```

### 最大值和最小值

```python
a=np.arange(20).reshape((4,5)) 
print(a)
print(a.min())
print(a.min(axis=0))# 各列的最小值，
print(a.min(axis=1))# 各行的最小值

# 最大值同理
```

### 最大值

```python
a=np.arange(20).reshape((4,5)) 
print(a)
print(a.sum())
print(a.sum(axis=0))# 每列求和
print(a.sum(axis=1))# 每行求和
```

### 加权平均数


```python
# 加权平均值=每个值*对应的权值求和，最后除以总的权值
a=np.array([80,90,95])
b=np.array([95,90,80])
weights=np.array([0.2,0.3,0.5])

print(np.average(a,weights=weights))
print(np.average(b,weights=weights))
```

在结构化数据类型创建可以使用元组形式进行定义。

- 每一个元组表示一个字段的，形式如(name,datatype,shape)

- 元组中的shape 字段是可选字段

- datatype可以定义为任何类型

```python
# 老师的数据类型有三个字段，name,age,salary .相当于c语言的结构体
# 表示单个字段的时候，需要使用元组进行表示
# name 字段指定shape为2的时候，多余的会进行截断
teacher = np.dtype([('name',np.str_,2),('age','i1'),('salary','f4')])
print(teacher)

ar2=np.array([
    ('w1',32,88998),
    ('ws',50,111),
    ('aaa',20,100)
],dtype=teacher)

print(ar2)
print(ar2[0]) # 访问数组中的第一个元素
print(ar2['name'])# 访问name字段的所有值
```
