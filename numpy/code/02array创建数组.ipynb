```python
# 
a = np.array([1,2,3,4,5])
# ba
b = np.array(a)
# abid
print('a:', id(a), '  b:', id(b))
print('ab')

# a =--
# b ---
b[0] = 10
print(a)


# 
c= a.copy()
print('a:',id(a),'c:',id(c))


```

```python
# ba
b[0] = 10
print('a:', a,'  b:', b)
print('='*10)
```

```python
# b=a 
a = np.array([1,2,3,4,5])
b = a
print(id(a))
print(id(b))

# ba
b[2]=999
print("b:")
print(id(a))
print(id(b))
print(a)
print(b)
```

```python
a = np.array([1,2,3,4,5])
# bcopyFasle
# 
b = np.array(a, copy=False)
# abid
print('a:', id(a), '  b:', id(b))
print('ab')
# abba
b[0] = 10
print('a:',a,'  b:',b)
```
