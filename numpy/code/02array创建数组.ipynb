### 设置copy参数，默认为True

#### 复习列表的深拷贝和浅拷贝

##### 1.浅拷贝

```python
c=[1,2,3,4,5]
d=c
d[2]=999
print(c)
```

##### 2.深拷贝

```python
# 方法一
c=[1,2,3,4,5]
d=c.copy();
d[2]=999;
print(d)
print(c)
```

---

```python
# 方法二
c=[1,2,3,4,5]
d=list(c)
d[2]=999;
print(d)
print(c)
```

默认情况下，传递一个数组作为参数，会复制一份，b指向新的内存地址空间。也就是深拷贝

```python
# 默认就是深拷贝
a = np.array([1,2,3,4,5])
# 定义b，复制a
b = np.array(a)
# 输出a和b的id
print('a:', id(a), '  b:', id(b))
print('以上看出a和b的内存地址')

# a =--复制
# b ---未复制
b[0] = 10
print(a)


# 第二种方法
c= a.copy()
print('a:',id(a),'c:',id(c))


```

```python
# 当修改b的元素时，a不会发生变化
b[0] = 10
print('a:', a,'  b:', b)
print('='*10)
```

```python
# b=a 这样是浅拷贝，
a = np.array([1,2,3,4,5])
b = a
print(id(a))
print(id(b))

# 当改变b的值的时候，也会改变a的值
b[2]=999
print("改变b的值之后:")
print(id(a))
print(id(b))
print(a)
print(b)
```

```python
a = np.array([1,2,3,4,5])
# 定义b，当设置copy参数为Fasle时，不会创建副本，
# 两个变量会指向相同的内容地址，没有创建新的对象
b = np.array(a, copy=False)
# 输出a和b的id
print('a:', id(a), '  b:', id(b))
print('以上看出a和b的内存地址')
# 由于a和b指向的是相同的内存地址，因此当修改b的元素时，a会发生变化
b[0] = 10
print('a:',a,'  b:',b)
```

### 设置ndmin指定数组的维度

```python
a = np.array([1,2,3])
b=np.array(a,ndmin=2)
print(b)
```

### subok参数，类型为bool值，默认False。为True，使用object的内部数据类型；False：使用object数组的数据类型。

```python
# 创建一个矩阵
a = np.mat([1,2,3,4])
# 输出为矩阵类型
print(type(a))

#既要复制一份副本，又要保持原类型
at = np.array(a,subok=True)
af = np.array(a)  # 默认为False
print('at,subok为True:',type(at))
print('af,subok为False:',type(af))
print(id(at),id(a))
```
