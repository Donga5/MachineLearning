## 案例：原生python代码和Numpy对比

### 引入numpy包

```python
import numpy as np # 一般别名就用np
```

```python
# 查看numpy的版本
np.__version__
```

### 1原生python实现

```python
def python_sum(n):
    '''
    原生Python实现两个数组的加法
    '''
    #  1~n的平方
    a=[i**2 for i in range(n)]
    #  1~n的立方
    b=[i**3 for i in range(n)]
    
    sum=[]
    for i in range(10):
        sum.append(a[i]+b[i])
    return sum
    
```

```python
# 调用函数
python_sum(10)
```

### 2numpy实现

```python
def numpy_sum(n):
    a=np.arange(n)**2
    b=np.arange(n)**3
    return a+b
```

```python
numpy_sum(10)
```

### 3.对比运行时间

```python
%timeit python_sum(1000)
```

```python
%timeit numpy_sum(1000)
```

#### 100W次数

```python
%timeit python_sum(1000000)
```

```python
%timeit numpy_sum(1000000)
```
