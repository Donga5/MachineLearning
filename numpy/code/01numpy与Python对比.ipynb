### 1python

```python
def python_sum(n):
    '''
    Python
    '''
    #  1~n
    a=[i**2 for i in range(n)]
    #  1~n
    b=[i**3 for i in range(n)]
    
    sum=[]
    for i in range(10):
        sum.append(a[i]+b[i])
    return sum
    
```

```python
# 
python_sum(10)
```

### 2numpy

```python
def numpy_sum(n):
    a=np.arange(n)**2
    b=np.arange(n)**3
    return a+b
```

```python
numpy_sum(10)
```
